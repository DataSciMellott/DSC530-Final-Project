```python
# Hypothesis 2
# Looking at this scatter plot, the distance from last transaction 
# does always mean fraud when the distance is large. Looks like distance away 
# could be an indicator of fraud, but not a certain representation
# Another factor that may make this more accurate is time between transactions with the distance. 
```

```python
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)

    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov

def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    meanx, varx = thinkstats2.MeanVar(xs)
    meany, vary = thinkstats2.MeanVar(ys)

    corr = Cov(xs, ys, meanx, meany) / np.sqrt(varx * vary)
    return corr

# Correlation Tests

print('Hypothesis 1')
print('used_chip to fraud')
print('Test 1: ', Corr(used_chip, fraud))
# low negative correlation between two variables in test1
print('')
print('Hypothesis 2')
print('dist_from_ltr to fraud')
print('Test 2: ', Corr(dist_from_ltr, fraud))
# Low positive correlation between two variables in test2
print('')
print('dist_from_hom to fraud')
print('Test 3: ', Corr(dist_from_home, fraud))
# low positive correlation
print('')
print('Hypothesis 3')
print('ratio_med_price to fraud')
print('Test 4: ', Corr(ratio_med_price, fraud))
# There is a medium correlation between the ratio medium purchase price and fraud. 
```
